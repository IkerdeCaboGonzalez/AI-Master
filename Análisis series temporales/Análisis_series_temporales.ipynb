```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
import seaborn as sns
import statsmodels.api as sm
import numpy as np
```

```python
# cargamos los datos

datos1=pd.read_csv('Sessions23sim.csv', sep=';')
datos2=pd.read_csv('Sessions23real.csv', sep=';')
```

```python
datos1.head()
```

```python
datos1.tail()
```

<h3> Serie 1 <h3>

```python
# Visualizamos la serie

serie1=pd.to_numeric(datos1['Serie1'].str.replace(',', '.'), errors='coerce')
serie1.dropna(inplace=True)
plt.figure(figsize=(10,5))
plt.plot(serie1)
plt.title('Serie 1')
plt.show()
```

```python
# Ploteo de la autocorrelación

plot_acf(serie1)
plt.show()
```

```python
plot_acf(serie1**2)
plt.title('Autocorrelación de Serie1 al cuadrado')
plt.show()
```

```python
plot_pacf(serie1)
plt.show()
```

```python
plot_pacf(serie1**2)
plt.title('Autocorrelación parcial de Serie1 al cuadrado')
plt.show()
```

```python
sns.kdeplot(serie1)
sns.histplot(serie1, bins=15, stat="density", color='skyblue')
plt.title('Gráfica de densidad de la serie 1')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
# Shapiro test

shapiro_test=stats.shapiro(serie1)
print(shapiro_test)
```

```python
# Dickey test

result = adfuller(serie1)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
media=serie1.mean()
desviacion=serie1.std()
print(media, desviacion)
```

```python
# Box test

box_test = sm.stats.acorr_ljungbox(serie1, lags=[25])
box_test
```
